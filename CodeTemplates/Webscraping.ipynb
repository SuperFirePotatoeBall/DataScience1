```python
#Main Scraping Challenge
#I am using the question:
#What is the most common word on Wikipedia?
#Technically, which word is the most common of the most articles
#I'm going to take a look at an idividual article, find the most common word, 
#and then see how many times a word appears as the most common for an article

#I will need to scrape though a lot of sites to get the data, I'm planning on like 100-1000
#Hopefully random articles, IDK if the API have access to this feature - It does!
```

```python
def commonWord(url):
    response = requests.get(url)
    if response.status_code != 200:
        print(response)
        return
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    
    article_text = ""
    for paragraph in soup.find_all('p'):
        article_text += paragraph.text
    words = article_text.split()
    word_count = Counter(words)
    most_common_word, frequency = word_count.most_common(1)[0]
    commonWords.append(most_common_word)
    global successfulURL
    successfulURL = successfulURL + 1
```

```python
def randomArticleURL():
    randomURL = "https://en.wikipedia.org/w/api.php?action=query&format=json&list=random&rnnamespace=0&rnlimit=1"
    response = requests.get(randomURL)
    data = response.json()
    random_article_title = data['query']['random'][0]['title']
    random_article_title = random_article_title.replace(" ", "_")
    content_url = f"https://en.wikipedia.org/wiki/{random_article_title}"
    print(random_article_title)
    
    return content_url
```

```python
#Main
commonWords = []
successfulURL = 0
while successfulURL <= 100:
    URL = randomArticleURL()
    commonWord(URL)
print(commonWords)
```
